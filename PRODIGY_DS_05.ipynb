Task 5 : Analyze traffic accident data to identify patterns related to road conditions. Visualize accident hotspots and contributing factors.


# Importing necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data loading and preprocessing

```python
accident_data = pd.read_csv('roadaccidentindia.csv')

# Handle missing values in the 'Count' column by replacing with drop 
accident_data['Count'].dropna
accident_data.head()
```

# Visualization 1: Top 10 Cities with Most Accidents

```python
city_accidents = accident_data.groupby('Million Plus Cities')['Count'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=city_accidents.head(10).index, y=city_accidents.head(10).values, palette='viridis')
plt.title('Top 10 Cities with Most Accidents', fontsize=16)
plt.xlabel('City', fontsize=12)
plt.ylabel('Total Accidents', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

# Visualization 2: Top 10 Causes of Accidents

```python
cause_accidents = accident_data.groupby('Cause category')['Count'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=cause_accidents.head(10).index, y=cause_accidents.head(10).values, palette='rocket')
plt.title('Top 10 Causes of Accidents', fontsize=16)
plt.xlabel('Cause Category', fontsize=12)
plt.ylabel('Total Accidents', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

# Visualization 3: Outcomes of Accidents

```python
outcome_accidents = accident_data.groupby('Outcome of Incident')['Count'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=outcome_accidents.index, y=outcome_accidents.values, palette='mako')
plt.title('Outcomes of Accidents', fontsize=16)
plt.xlabel('Outcome', fontsize=12)
plt.ylabel('Total Count', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```
